##Customer Segmentation using K-Means

```python
X = df[
[
'Income',
'MntTotal',
'TotalPurchases',
'Recency',
'Age'
]
]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='Income',
    y='MntTotal',
    hue='Cluster',
    palette='Set1'
)

plt.title("Customer Segments")
plt.show()
```

```python
df.groupby('Cluster')[
    ['Income','MntTotal','TotalPurchases','Recency','Age']
].mean()
```

##Visualization

```python
cluster_counts = df['Cluster'].value_counts().sort_index()

sns.barplot(
    x=cluster_counts.index,
    y=cluster_counts.values
)

plt.title('Number of Customers in Each Segment')
plt.xlabel('Cluster')
plt.ylabel('Customer Count')
plt.show()
```

```python
cluster_income = df.groupby('Cluster')['Income'].mean()

cluster_income.plot(kind='bar')

plt.title('Average Income by Customer Segment')
plt.ylabel('Average Income')
plt.show()
```

```python
cluster_spending = df.groupby('Cluster')['MntTotal'].mean()

cluster_spending.plot(kind='bar')

plt.title('Average Spending by Customer Segment')
plt.ylabel('Average Spending')
plt.show()
```

## Key Insights

1. Four distinct customer segments were identified using K-Means clustering.
2. Cluster 3 consists of premium customers with the highest income and spending levels.
3. Cluster 0 contains loyal high-value customers who make frequent purchases.
4. Cluster 2 represents budget-conscious customers with lower spending patterns.
5. Cluster 1 consists of inactive customers who may require re-engagement strategies.

## Recommendations

1. Provide VIP rewards and exclusive offers to premium customers (Cluster 3).

3. Offer discounts and value-based promotions to budget customers (Cluster 2).
4. Run re-engagement campaigns for inactive customers (Cluster 1).
