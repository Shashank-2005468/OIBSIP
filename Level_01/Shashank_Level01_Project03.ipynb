#Importing required python libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#Loading dataset

```python
df=pd.read_csv("/content/AB_NYC_2019.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

##**Handle Missing Values**

```python
df["name"].fillna("Unknown", inplace=True)
```

```python
df["host_name"].fillna("Unknown", inplace=True)
```

```python
df.isnull().sum()
```

```python
df['last_review']=pd.to_datetime(df['last_review'])
```

```python
df['last_review'].fillna(df['last_review'].mode()[0],inplace=True)
```

```python
df['reviews_per_month'].fillna(df['reviews_per_month'].median(), inplace=True)
```

```python
df.isnull().sum()
```

## Missing Data Handling

Missing values were identified in the columns name, host_name, last_review, and reviews_per_month. Appropriate imputation techniques were applied to handle the missing data. After preprocessing, no missing values remained in the dataset.

```python
df.duplicated().sum()
```

## Duplicate Removal

The dataset was checked for duplicate records using the duplicated() function. No duplicate records were found in the dataset, indicating that each listing is unique.

##Loading Dataset

```python
df=pd.read_csv('/content/INvideos.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['description'].fillna('No Description',inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

## Duplicate Removal

The dataset contained 4,263 duplicate records. These duplicate entries were identified using the duplicated() function and removed using the drop_duplicates() method. This improved the uniqueness and reliability of the dataset.

```python
df['trending_date'] = pd.to_datetime(
    df['trending_date'],
    format='%y.%d.%m'
)
```

```python
df['publish_time'] = pd.to_datetime(
    df['publish_time']
)
```

```python
df.columns = df.columns.str.strip()
```

```python
df.head()
```

```python
df['title'] = df['title'].str.strip()
```

```python
df['tags'] = df['tags'].str.strip()

df['description'] = df['description'].str.strip()
```

#Visualization

```python
sns.boxplot(x=df['views'])
```

```python
df['channel_title'] = df['channel_title'].str.strip()
df['channel_title'] = df['channel_title'].str.replace('\n', ' ', regex=False)
```

```python
print(df['channel_title'].head())
```

##Result
1. The dataset contained 561 missing values in the description column, which were handled appropriately.
2. A total of 4263 duplicate records were identified and removed.
3. Date columns were converted into a standardized datetime format.
4. Text columns were cleaned by removing unnecessary whitespace and formatting inconsistencies.
